The following cell will calculate the sin(x) for a given number:

```python
import numpy as np

variablex = float(input("Please enter a value for x:"))
print("sin(x) is " + str(f"{np.sin(variablex):.5f}"))
```

The following cell will calculate the cos(x) for a given number:

```python
variablex = float(input("Please enter a value for x:"))
print("cos(x) is " + str(f"{np.cos(variablex):.5f}"))
```

The following cell will tabulate x between 0 and 2pi with a thousand entries:

```python
from astropy.table import Table

variablex = np.arange(0.0062831853, ((2*np.pi)-0.00627690212), 0.00627690212)
sin_variablex = np.sin(variablex)
cos_variablex = np.cos(variablex)

sincos_table = Table()
sincos_table["x"] = variablex
sincos_table["sin(x)"] = sin_variablex
sincos_table["cos(x)"] = cos_variablex

sincos_table["x"].format = "{:.3f}"
sincos_table["sin(x)"].format = "{:.3f}"
sincos_table["cos(x)"].format = "{:.3f}"
```

The following cell will take the first 10 rows from the above tabulated table and print them:

```python
first10_table = Table(names=sincos_table.colnames)

first10_table["x"].format = "{:.3f}"
first10_table["sin(x)"].format = "{:.3f}"
first10_table["cos(x)"].format = "{:.3f}"

for i in range(10):
    first10_table.add_row(sincos_table[i])

print(first10_table)
```
