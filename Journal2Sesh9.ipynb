```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
figures = np.random.uniform(low=0, high=1, size=1000)
```

```python
plt.hist(figures, bins=100)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show

fname='journal2sesh9.png'
plt.savefig(fname,bbox_inches='tight',dpi=400)
```
